```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

%matplotlib inline
```

```python
no_block_readings = []
block_readings = []
torque_names = None
for i in range(1, 6):
    readings = []
    with open('../traj/traj_no_block_%d'%i, 'r') as f:
        reader = csv.DictReader(f)
        torque_names = reader.fieldnames
        for l in reader:
            readings.append(l.values())
    no_block_readings.append(np.asarray(readings))

for i in range(1, 6):
    readings = []
    with open('../traj/traj_sauce_block_%d'%i, 'r') as f:
        reader = csv.DictReader(f)
        for l in reader:
            readings.append(l.values())
    block_readings.append(np.asarray(readings))

no_block_data = np.asarray(no_block_readings)
block_data = np.asarray(block_readings)
```

```python
for i, name in enumerate(torque_names):
    plt.figure()
    plt.title(name)
    red_patch = mpatches.Patch(color='red', label='Block')
    blue_patch = mpatches.Patch(color='blue', label='No block')
    for readings in no_block_data:
        try:
            plt.plot(readings[:, i], color="b")
        except:
            pass
    for readings in block_data:
        try:
            plt.plot(readings[:, i], color="r")
        except:
            pass
    plt.legend(handles=[red_patch, blue_patch])
    plt.savefig('{}_readings.pdf'.format(name))
```
